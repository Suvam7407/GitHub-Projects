```python
# spliting the every rows using split method and add a new column ,name is split
df['split']=df['messages'].apply(lambda x: x.split(" "))
```

```python
# top 5 rows 
df.head()
```

```python
# take 2 list a,n for append keyword and line of the messages and append the keyword with line numbers into differnt list a and n
a=[]
n=[]
# for w in kw:
#   for i in range(df.shape[0]):
#     if w in df["split"][i]:
#       print(w,i)
#       a.append(w)
#       n.append(i)
for k in range(len(kw)):
  if len(kw[k].split())==1:
    for i in range(df.shape[0]):
      if kw[k].split()[0] in df['split'][i]:
        print(kw[k],i)
        a.append(kw[k])
        n.append(i)
  else:
    for i in range(df.shape[0]):
      if kw[k].split()[0] in df['split'][i] and kw[k].split()[1] in df['split'][i]:
        print(kw[k],i)
        a.append(kw[k])
        n.append(i)
      
     
     

```

```python
# create a dict. 
df1={
    'Keywords':a,
     "Numbers":n
}
```

```python
# convert a dict to dataframe .there have 2 columns name keywords and numbers
df_all=pd.DataFrame(df1,columns=['Keywords','Numbers'])
df_all
```

```python
# using group by find out the every keywords frequency 
df_all.groupby(["Keywords"])['Numbers'].count().nlargest(14)


```
